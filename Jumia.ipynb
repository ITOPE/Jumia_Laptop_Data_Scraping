```python
import pandas as pd
from bs4 import BeautifulSoup as Soup
import requests as Re
import numpy as np
import csv
```

```python
Jumia=[]
for x in range(1,51):
    url=f"https://www.jumia.com.ng/catalog/?q=laptops&page={x}#catalog-listing"
    html=Re.get(url).text
    html_data=Soup(html, "html.parser")
    data=html_data.find_all("div",{"info"})
    
    for Laptop_data in data:
        Name=Laptop_data.find("h3",{"class":"name"}).text
        Price=Laptop_data.find("div", {"class":"prc"}).text
        Price=Price[1:].replace(",","")
        try:
            Old_Price=(Laptop_data.find("div", {"class":"old"}).text)[1:].replace(",","")
        except:
            Old_Price=None
        try:
            Discount=Laptop_data.find("div", {"class":"bdg _dsct _sm"}).text

        except:
            Discount=None
        try:
            Rating=(Laptop_data.find("div", {"class":"stars _s"}).text)[:3].replace("o","")
        except:
            Rating=None
        
        Jumia.append([Name, Price, Old_Price, Discount, Rating])
    
```

```python
df=pd.DataFrame(Jumia, columns=["Name", "Price", "Old Price", "%Discount", "Rating"])
df
```

```python
df.to_csv("Jumia Laptop Data.csv")
```
